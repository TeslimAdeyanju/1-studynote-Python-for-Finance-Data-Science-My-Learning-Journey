```python
import arff
import pandas as pd

# Load ARFF file
with open('Currency-Exchange-Rates.arff') as file:
    dataset = arff.load(file)

# Extract column names from the attributes
column_names = [attr[0] for attr in dataset['attributes']]

# Convert to pandas DataFrame with column names
fx_data = pd.DataFrame(dataset['data'], columns=column_names)

# Display the DataFrame

fx_data.head()
```

```python
# select the numeric data group 
numeric_data_columns = fx_data.select_dtypes(include=['float64'])
numeric_data_columns
```

```python
# checking for the missing data 
fx_data.isnull().sum()
```

```python
# Cleaning column headers 
def clean_column_names(df):
    # ensure columns is of string type
    df.columns = df.columns.astype(str)
    # strip of any leading and trailing spaces 
    df.columns = df.columns.str.strip()
    # convert all columns headers to lower case 
    df.columns = df.columns.str.lower()
    return df


# calling the function 
clean_column_names(fx_data) 
```

```python
# Filling the missing values with the median of the column
def fill_missing_values(df, column_names):
    median_value = df[column_names].median()
    df[column_names] = df[column_names].fillna(median_value)
    return df

for column_names in numeric_data_columns:
    clean_data_for_use = fill_missing_values(numeric_data_columns, column_names)

clean_data = numeric_data_columns
clean_data
                          
                          
```
