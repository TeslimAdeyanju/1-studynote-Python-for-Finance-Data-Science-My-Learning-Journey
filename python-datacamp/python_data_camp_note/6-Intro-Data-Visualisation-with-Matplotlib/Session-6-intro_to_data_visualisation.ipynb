# Module 1: Introduction to Matplotlib

The first step in creating a data visualization is to set up the canvas. 

In Matplotlib, the `canvas` is known as a `figure`. The figure is the overall window or page that everything is drawn on. Within the figure, we can contain multiple plots. These plots are known as `axes`. The axes are the actual plots. They will contain the ticks, labels, etc. that make up the plot.

In this mission, we will learn how to create a figure and add axes to it. We will also learn how to customize the appearance of the axes.

The syntax for creating a figure is simple:

```python

import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data on the axis
ax.plot(df[columns], df[columns]) 

# Add labels and title
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Plot Title')

# Show legend (optional)
ax.legend()

# Display grid (optional)
ax.grid(True)

# Show plot
plt.show()

```


````python
# This is just a simple plot of the data
fig, ax= plt.subplots()
plt.show()
````

````python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color) 
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color) 
    axes.tick_params('y', colors=color)
````

````python
fig, ax = plt.subplots()

plot_timeseries(ax, btc_usd_2012.index, btc_usd_2012['close'], 'blue', 'Time', 'F' )
````

### 2.0 Plotting time-series with different variables

````python
# combining two plots of six months record and the whole year record
"""
This script combines two plots: a six-month record and a whole year record of Bitcoin prices.

Functions:
    plt.figure(figsize=(20, 6)): Sets the figure size for the plot.
    plt.plot(btc_usd_2012.index, btc_usd_2012['close']): Plots the closing prices of Bitcoin for the whole year.
    ax.plot(btc_usd_2012.index, btc_usd_2012['profit']): Plots the profit of Bitcoin for the whole year.
    plt.plot(six_mothns_record.index, six_mothns_record['close'], color='red'): Plots the closing prices of Bitcoin for six months in red color.
    plt.show(): Displays the combined plot.
"""
plt.figure(figsize=(20, 6))

plt.plot(btc_usd_2012.index, btc_usd_2012['close'])
plt.plot(btc_usd_2012.index, btc_usd_2012['profit'])
plt.plot(six_mothns_record.index, six_mothns_record['close'], color='red')
plt.show()
````

````python
# using the twin axis to plot the two data
fig, ax = plt.subplots(figsize=(20, 6), nrows=1, ncols=1)

ax.plot(btc_usd_2012.index, btc_usd_2012['profit'])
ax.set_xlabel('date')
ax.set_ylabel('frequencies')

ax2 = ax.twinx()


ax2.plot(six_mothns_record.index, six_mothns_record['close'], color='red')
ax2.set_ylabel('Six month indicator')
plt.grid(True)
plt.ion()

plt.show()
````
