```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data
data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'SubCategory': ['X', 'Y', 'X', 'Y', 'Y', 'X', 'X', 'Y', 'Y']
}
df = pd.DataFrame(data)

# Count Plot with hue
sns.countplot(data=df, x='Category', hue='SubCategory')
plt.title("Count Plot with hue")
plt.xlabel("Category")
plt.ylabel("Count")
plt.legend(title="SubCategory")
plt.show()
```

```python
sns.countplot(data=df, x='Category', order=['C', 'B', 'A'])
plt.title("Ordered Count Plot")
plt.show()
```

```python
sns.countplot(
    data=df,
    x='Category',
    palette='viridis',
    order=['C', 'B', 'A'],
    hue='SubCategory'
)
plt.show()

```

```python
sns.countplot(data=df, x='Category', order=['C', 'B', 'A'])
plt.title("Ordered Count Plot")
plt.show()
```

```python
sns.countplot(
    data=df,
    x='Category',
    hue='SubCategory',
    hue_order=['Y', 'X']
)
plt.title("Count Plot with Hue Order")
plt.legend(title="SubCategory")
plt.show()

```

```python
sns.countplot(data=df, x='Category', orient='v')  # Vertical
plt.title("Vertical Orientation")
plt.show()

sns.countplot(data=df, y='Category', orient='h')  # Horizontal
plt.title("Horizontal Orientation")
plt.show()

```

```python
sns.countplot(data=df, x='Category', palette='Spectral', hue='SubCategory')
plt.title("Count Plot with Custom Palette")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Display first few rows of the dataset
titanic.head()
```

```python
sns.scatterplot(data=titanic, x='age', y='fare')
plt.title("Basic Scatter Plot")
plt.show()
```

```python
sns.scatterplot(data=titanic, x='age', y='fare', size='sibsp', sizes=(20, 200))
plt.title("Scatter Plot with Variable Sizes")
plt.legend(title="Number of Siblings/Spouses")
plt.show()
```

```python
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived', palette='dark')
plt.title("Scatter Plot with Custom Palette")
plt.legend(title="Survived")
plt.show()
```

```python
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived', hue_order=[1, 0], palette='dark')
plt.title("Scatter Plot with Ordered Hue")
plt.legend(title="Survived")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = sns.load_dataset('tips')
data.head()
```

```python
sns.catplot(data=data, 
            x='time', 
            y='total_bill', 
            kind='box', 
            fliersize=0)
            
            
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
sns.catplot(data=data, 
            x='time', 
            y='total_bill', 
            kind='box')
            
            
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            kind='strip', 
            jitter=True)
            
plt.title("Strip Plot of Total Bill by Day")
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            kind='swarm', 
            hue='sex')
            
plt.title("Swarm Plot of Total Bill by Day with Hue")
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            kind='box', 
            hue='smoker',
              fliersize=0)
            
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            kind='violin', 
            hue='smoker', 
            split=True)
            
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            kind='boxen')
            
plt.title("Boxen Plot of Total Bill by Day")
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            kind='bar', 
            hue='time', 
            errorbar='sd')
plt.title("Bar Plot of Total Bill by Day")
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            kind='point', 
            hue='time')
plt.title("Point Plot of Total Bill by Day")
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            kind='count', 
            hue='sex')
plt.title("Count Plot of Days with Hue")
plt.show()
```

```python
sns.catplot(
    data=data,
    x='day',
    y='total_bill',
    hue='smoker',
    kind='bar',
    order=['Sat', 'Sun', 'Thur', 'Fri'],
    hue_order=['Yes', 'No']
)
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            kind='bar', 
            hue='smoker', 
            palette='coolwarm')
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            kind='bar', 
            height=5, 
            aspect=1.1)
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            col='sex', 
            kind='bar', 
            margin_titles=True)
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            kind='violin', 
            hue='smoker', 
            legend_out=False)
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            col='time', 
            kind='bar', 
            sharey=False)
plt.show()
```

```python
sns.catplot(data=data, 
            x='day', 
            y='total_bill', 
            hue='sex', 
            kind='swarm', 
            dodge=True)
plt.show()
```

```python
sns.catplot(
    data=data,
    x='day',
    y='total_bill',
    hue='smoker',
    col='sex',
    row='time',
    kind='bar',
    height=6,
    aspect=1.2,
    palette='muted',
    margin_titles=True,
    sharey=False
)
plt.show()
```
