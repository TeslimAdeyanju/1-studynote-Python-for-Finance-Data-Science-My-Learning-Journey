```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import seaborn as sns

# Load the dataset
penguins = sns.load_dataset("penguins")

# Create the figure and GridSpec
fig = plt.figure(figsize=(12, 8))
grid = GridSpec(3, 3, figure=fig)

# Add subplots with custom sizes
ax1 = fig.add_subplot(grid[0, 0])  # Top-left cell
sns.histplot(penguins['flipper_length_mm'], ax=ax1, kde=True, color='blue')
ax1.set_title('Flipper Length')

ax2 = fig.add_subplot(grid[0, 1:])  # Top-right two cells
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', ax=ax2)
ax2.set_title('Bill Length vs. Depth')

ax3 = fig.add_subplot(grid[1:, 0])  # Bottom-left two rows
sns.boxplot(data=penguins, x='species', y='body_mass_g', ax=ax3, palette='Set2')
ax3.set_title('Body Mass by Species')

ax4 = fig.add_subplot(grid[1:, 1:])  # Bottom-right section
sns.heatmap(penguins.corr(numeric_only=True), annot=True, cmap='coolwarm', ax=ax4)
ax4.set_title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import seaborn as sns

# Load the dataset
penguins = sns.load_dataset("penguins")

# Create the figure
fig = plt.figure(figsize=(14, 8))
grid = GridSpec(3, 4, figure=fig)

# GridSpec: Custom-sized plots
ax1 = fig.add_subplot(grid[0, 0:2])  # Top-left two columns
sns.histplot(penguins['flipper_length_mm'], ax=ax1, kde=True, color='blue')
ax1.set_title('Flipper Length')

ax2 = fig.add_subplot(grid[0, 2:])  # Top-right two columns
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', ax=ax2)
ax2.set_title('Bill Length vs. Depth')

# plt.subplots: Uniform grid below
axes = [fig.add_subplot(grid[i, j]) for i in range(1, 3) for j in range(4)]
for i, ax in enumerate(axes):
    sns.boxplot(data=penguins, x='species', y='body_mass_g', ax=ax, palette='Set2')
    ax.set_title(f'Subplot {i+1}')

plt.tight_layout()
plt.show()
```
