````python
# Define a function to clean numeric data

def clean_numeric_data(df, column_name):
    # Convert the entire column to string first
    df[column_name] = df[column_name].astype(str)
    # Remove commas
    df[column_name] = df[column_name].str.replace(',', '', regex=False)
    # Strip leading and trailing spaces
    df[column_name] = df[column_name].str.strip()
    # Replace problematic strings with '0.0' (or use np.nan if you prefer)
    df[column_name] = df[column_name].replace(['#REF!', 'Nil', 'NIl', 'NIl!'], '0.0')
    # Convert the data into float, coercing errors to NaN
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
    return df

# Assuming df_financial_numerical is defined and has numeric data with potential issues
# Convert the numerical data to float
for column in df_financial_numerical.columns:
    df_financial_numerical = clean_numeric_data(df_financial_numerical, column)

# Checking the first few rows to verify the changes
df_financial_numerical.head()
````

1. **First Line:**
   ```python
   df_presser_note_4.loc[("alidu", "suit"):("dominc", "buba"), "jan":"oct"]
   ```
   - **Selection Criteria:** This line uses a tuple for the row index selection, implying that the DataFrame has a MultiIndex (hierarchical index) for its rows.
   - **Implications:** It selects rows from the MultiIndex starting from the tuple `("alidu", "suit")` to the tuple `("dominc", "buba")` and columns from "jan" to "oct". This is useful when you have a DataFrame with multiple levels of indexing and you want to slice based on those levels.

2. **Second Line:**
   ```python
   df_presser_note_4.loc["alidu":"dominc", "jan":"oct"]
   ```
   - **Selection Criteria:** This line uses simple string labels for the row index selection, implying that the DataFrame has a single-level index for its rows.
   - **Implications:** It selects rows starting from "alidu" to "dominc" and columns from "jan" to "oct". This is a straightforward slicing operation for a DataFrame with a single-level index.

### Summary of Differences and Implications:
- **Index Type:** The first line assumes a MultiIndex, while the second line assumes a single-level index.
- **Selection Range:** The first line selects a range of rows based on tuples in a MultiIndex, whereas the second line selects a range of rows based on single-level index labels.
- **Use Case:** Use the first line when dealing with hierarchical indexing and the second line for simpler, single-level indexing.

#### 1. Creating a pivot table
____

The syntax for a creating a pivot table is as follows:

`df.pivot_table(values='column_name1', index='column_name2', columns='column_name3', aggfunc='function')`

where 
- `values` is the column to aggregate,  
- `index` is the column to group by,
- `columns` is the column to pivot, and
- `aggfunc` is the aggregation function to apply.


````python
# Pivot the data the data frame
presser_pivot_jan = df_presser.pivot_table(values='jan', index='name', columns='mode', fill_value=0)
presser_pivot_jan
````

````python
presser_pivot_feb = df_presser.pivot_table(values='feb', index='name', columns='mode', fill_value=0)
presser_pivot_feb
````

#### 2. Slingcing a pivot table with .loc[]
____

Just  like the dataframe, we cab combine the pivot table with the `df.loc[]` method to subset the pivot table. The syntax is as follows:

`df.pivot_table(values='column_name1', index='column_name2', columns='column_name3', aggfunc='function').loc["row_name1":"row_name2", "column_name1":"column_name2"]`



````python

presser_pivot_feb.loc["alidu":"okon"]
````

### 3. Axis parameter in pivot_table()
____

We can indicate the axis along which the pivot table should be created by passing the `axis` parameter to the `pivot_table()` method. The `axis` parameter can take the values `0` or `1`, where `0` indicates that the pivot table should be created along the index, and `1` indicates that the pivot table should be created along the columns.

````python
presser_pivot_feb.sum(axis=1)
````

````python
presser_pivot_feb.sum(axis=0)
````

````python
presser_pivot_feb.mean(axis=1)
````

You can manually compute the histogram and then plot it using [`plt.bar`]

### Missing Values
____

- It is often a pratice that the data we have have some missing values, and first, we need to check our data if it contains any missing values. In the pdandas dataframe, missing values are represented by `NaN` or `None`.

- if we get our dataset, first use the df.isna().sum() to check for missing values in the dataset. This only returns True or False, and to get the sum of the missing values in each column, we can use the df.isna().sum() method.

- Note that the `df.isna()` method returns a DataFrame of the same shape as the original DataFrame, where each entry is a boolean value indicating whether the corresponding entry in the original DataFrame is missing or not. The `sum()` method is then used to count the number of missing values in each column.

-Note that df.isnull() is an alias for df.isna(), and both methods can be used interchangeably as df.isnull().sum() or df.isna().sum().



````python
import pandas as pd
file_path = r'/Users/teslim/OneDrive/3_adult_ws_1.csv'

data = pd.read_csv(file_path)
data.head()
````

````python
# data information
data.info()
````

````python
# Detecting missing values
data.isna()
````

````python
# We can also check if the dataframe comntains any missing values
data.isna().any()
````

````python
# counting missing values 
data.isna().sum()
````

````python
# We can use plot to visualize the missing values
data.isna().sum().plot(kind='bar')
````

After the detection of missing values, we can either drop the missing values or fill them with a specific value.

- droping missing values: we can use the `df.dropna()` method to drop missing values from the DataFrame. The `df.dropna()` method removes rows with missing values from the DataFrame. The syntax is as follows:

`df.dropna()`

- filling missing values: we can use the `df.fillna()` method to fill missing values in the DataFrame. The `df.fillna()` method fills missing values in the DataFrame with a specified value. The syntax is as follows:

`df.fillna(value)`


````python
# this will drop all the rows with missing values
data = data.dropna()
````
