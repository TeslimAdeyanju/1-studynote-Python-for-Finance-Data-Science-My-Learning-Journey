```python
twenty_fifth = df1['Retail_Price'].quantile(0.25)
median = df1['Retail_Price'].quantile(0.5)
seventy_fifth = df1['Retail_Price'].quantile(0.75)
maximum = df1['Retail_Price'].max()
```

```python
label = ["Economy", "Premium Economy", "Business Class", "First class"]
```

```python
bins = [0,twenty_fifth, median, seventy_fifth, maximum]
```

```python
df1['info'] = pd.cut(df1['Retail_Price'], bins=bins, labels=label)
df1['info'].value_counts()
```
