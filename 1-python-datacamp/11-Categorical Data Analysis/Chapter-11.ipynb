```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rf= pd.read_csv('lasvegas_tripadvisor.csv')
rf
```

```python
def clean_header(df):
    df.columns = df.columns.str.strip()
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(r'\s+',  '_')
    df.columns = df.columns.str.replace(r'_+' ,  '_')
    return df
    
    
# calling the function 
clean_header(rf)
```

```python
def clean_content_column(df, col):
    df[col] = df[col].astype(str)
    df[col] = df[col].str.lower()
    df[col] = df[col].str.replace(" ", "-")
    return df
```

```python
for col in rf:
    clean = clean_content_column(rf, col)
```
