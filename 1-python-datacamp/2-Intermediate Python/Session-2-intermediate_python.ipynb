```python
import numpy as np

# Given returns
returns = [-0.016, 0.317, 0.074, -0.126, -0.124]

# Convert to decimal multipliers by adding 1
multipliers = [1 + r for r in returns]

# Calculate the product of the multipliers
product = np.prod(multipliers)

# Calculate the geometric mean by taking the n-th root (n is the number of returns)
geometric_mean = product**(1/len(multipliers))

# Convert back to original form by subtracting 1
geometric_mean_percentage = (geometric_mean - 1) * 100

# Print the result
print(f"The geometric mean of the given returns is approximately {geometric_mean_percentage:.2f}%")
```

```python
# NumPy is imported; seed is set
np.random.seed(123)
# Initialize all_walks (don't change this line)
all_walks = []

# Simulate random walk five times
for i in range(5):

    # Code from before
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
print(all_walks)
```

```python
bytes1 = b'\x04\x00'
int1 = int.from_bytes(bytes1, byteorder='big')  # Big-endian
print(int1)

bytes2 = b'\x00\x04'
int2 = int.from_bytes(bytes2, byteorder='little')  # Little-endian
print(int2)
```
