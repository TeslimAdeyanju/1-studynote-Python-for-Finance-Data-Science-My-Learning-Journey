```python
# clean the column_headers

def clean_columns_header(df):
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.strip()
    df.columns = df.columns.str.replace(" ", "_")
    return df


    
```

```python
## clean the content of the column

def clean_content_column(df,columns_name):
    df[columns_name] =  df[columns_name].astype(str)
    df[columns_name] =  df[columns_name].str.lower()
    df[columns_name] =  df[columns_name].str.strip()
    df[columns_name] =  df[columns_name].str.replace(" ", "")
    return df
```

```python
for columns_name in categorical_data:
    clean_data = clean_content_column(df,columns_name)
```
